```python
datanames1 = ['adult',
'agaricus_lepiota',
'audit',
'australian',
'breast',
'breast_cancer',
'breast_cancer_wisconsin',
'buggyCrx',
'chess',
'churn',
'clean1',
'clean2',
'cleve',
'coil2000',
'colic',
'contraceptive',
'credit_a',
'credit_g',
'crx',
'dis',
'flare',
'german',
'glass2',
'heart',
'heart_c',
'heart_h',
'Hill_Valley_with_noise',
'Hill_Valley_without_noise',
'horse_colic',
'hypothyroid',
'ionosphere',
'kr_vs_kp',
'mofn_3_7_10',
'mushroom',
'phoneme',
'pima',
'prnn_crabs',
'ring',
'sonar',
'spambase',
'spectf',
'tokyo1',
'wdbc',
'magic',
'diabetes',
"profb",
'pima']

datanames1 = pd.DataFrame(datanames1,columns = ['dataset_names'])
datanames1
```

```python
pmlb_data = pd.read_csv('https://raw.githubusercontent.com/EpistasisLab/pmlb/master/pmlb/all_summary_stats.tsv', sep='\t')
pmlb_data['percent_binary'] = pmlb_data['n_binary_features']/pmlb_data['n_features']
#pmlb_data
binary = pd.merge(datanames1, pmlb_data[['dataset','percent_binary']], left_on = 'dataset_names',right_on = 'dataset',how = 'left')
binary = binary.fillna(np.median(0)).drop('dataset',axis = 1)

```

```python
diff = []
for name in datanames:
    path = 'bagboostresults/' + name
    print(path)
    final_result_agg = pd.read_csv(path)

    final_result_agg.drop(0,axis = 0,inplace = True)
    try:
        final_result_agg.columns = ['nonzero','polish_mean','polish_std','base_mean','base_std','full_mean','full_std']
    except:
        print(name + 'Error')
        final_result_agg.columns = ['nonzero','polish_mean','polish_std','base_mean','base_std']
    final_result_agg = final_result_agg.astype(float)
    temp = final_result_agg.head(10)

    diff.append(np.mean(temp['polish_mean'] - temp['base_mean']))
```

```python
pmlb_data = binary.copy()
pmlb_data['perf_diff'] = diff
pmlb_data
```

```python
plt.hist(diff, bins = 10)
plt.xlabel('Difference in ROC AUC over first 10 Features')
```

```python
import matplotlib
plt.figure(figsize = (12,8))

cm = plt.cm.get_cmap('Greys_r')

colors = []
n, bins, patches = plt.hist(pmlb_data['perf_diff'],bins =10)
plt.xlabel('Mean Difference In AUC between ' r'\textsc{ControlBurn}' +' and Random Forest Baseline' )

for i in range(len(bins)-1):
    temp = pmlb_data[pmlb_data['perf_diff'] >= bins[i]]
    temp = temp[temp['perf_diff'] < bins[i+1]]
    colors.append(np.mean(temp['percent_binary']))

for c, p in zip(colors, patches):
    plt.setp(p, 'facecolor', cm(c))
    
norm = matplotlib.colors.Normalize(vmin=0, vmax=1, clip=False)
cbar = plt.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cm))

ax = cbar.ax
ax.text(3.2,0.2,'Percentage of Features Binary',rotation=90)
#cbar.set_label('Percentage of Features Boolean')
plt.ylabel('Counts')
plt.title('PMLB Benchmark Data Results')
#plt.savefig('pmlbhist.pdf',bbox_inches = 'tight',
#    pad_inches = 0)
```
